```python
#join_date is object (string), but it should be datetime.
#age, salary, score may be float even though they could be int.
```

```python
#Convert columns to numeric types
#Convert column to integer, handling NaNs:
df["age"] = df["age"].astype("Int64")  # Pandas nullable integer
df["score"] = df["score"].astype("Int64")
```

```python
int64 not handle none 
Int64 can handle none and convert into NAN
object can be convert into string (none converted into <na>)
```

```python
df
```

```python
df.dtypes
```

```python
The column contains integers + maybe some NaNs use  Int64
#when  you need to convert a messy column into numbers. use to_numeric with error='coerce'
```

```python
pd.to_numeric()
if the column contains strings, mixed types, or invalid numbers.
Converts a  df["salary"] to numeric type (int or float).
errors="coerce"

Invalid or unconvertible values (like "abc", "NaN", or "one million") are replaced with NaN instead of raising an error.
```

```python
df["salary"] = pd.to_numeric(df["salary"], errors="coerce") 
#Use it when you need to convert a messy column into numbers.
```

```python
#Convert text columns to categorical
#For columns with limited categories like gender or department:

df["gender"] = df["gender"].astype("category")
df["department"] = df["department"].astype("category")

#Benefits:
Reduces memory usage
Faster groupby / filtering
```

# Convert date columns

```python
default MM-DD-YYYY.
Pandas will interpret dates in DD-MM-YYYY when use dayfirst=True
```

```python
df["join_date"] = pd.to_datetime(df["join_date"], errors="coerce", dayfirst=True)

#errors="coerce" → invalid dates become NaT (null datetime)
#dayfirst=True → interprets 15-07-2019 as 15th July
```
