```python
s1 = pd.Series([10, 20, 30])
s1
```

```python
s2 = pd.Series([1, 2.2, 3,1]) # convert all into float
s3 = pd.Series(['a', 'b', 'c',None,1,-1,2.3])
s4 = pd.Series([True, False, True])
s5 = pd.Series([10, 20, 30,None])

print(s1.dtype)  # int64
print(s2.dtype)  # float64
print(s3.dtype)  # object heterogenous
print(s4.dtype)  # bool
print(s5.dtype) # float64
print(s2)
print(s1)
print(s5)
```

# Series vs List

```python
           Feature             Python List                     pandas Series
0            Index  Only integer positions             Custom index (labels)
1            Speed   Slower for large data    Faster (uses NumPy internally)
2  Math operations  Not supported directly      Fully vectorized (fast math)
3   Missing values              No support             NaN handling built-in
4          Methods                Very few  Many (mean, sum, describe, etc.)
5      Integration      General Python use            Best for data analysis
```

```python
lst = [10, 20, 30]
s = pd.Series(lst)

print(s * 2)
# Series supports math → list does not.
```

# boolean indexing

```python
#used in filtering the data 
s = pd.Series([10, 20, 30, 40, 50])

# Filter values greater than 25
filtered = s[s > 25]
print(filtered)
```

```python
s = pd.Series([10, 20, 30, 40, 50])

# Filter values greater than 25
s[s > 25]
```

```python
# Values > 20 AND < 50
filtered = s[(s > 20) & (s < 50)]
print(filtered)
```

```python
## Values < 20 OR > 40
filtered = s < 20
print(filtered)
s[filtered]
```

```python
## Keep only 20 and 50
filtered = s[s.isin([20, 50])]
print(filtered)
```

```python
# Keep values NOT 20 or 50
filtered = s[~s.isin([20, 50])]
print(filtered)
```
