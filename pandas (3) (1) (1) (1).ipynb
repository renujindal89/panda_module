# outliers 

```python
import numpy as np
import pandas as pd
```

```python
#IQR Method (Interquartile Range)  MOST COMMON
Outliers are values outside:

[ Q1  1.5*IQR ,  Q3 + 1.5*IQR ]
```

```python
df = pd.DataFrame({
    'height': [150, 152, 153, 151, 149, 155, 300, 148, 152, 154],
    'weight': [50, 52, 53, 49, 51, 55, 200, 48, 52, 54]
})
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,2))
plt.boxplot(df['height'], vert=False)
plt.title("Height Boxplot")
plt.grid(True)
plt.show()
```

```python
Five-Number Summary
Minimum (not counting outliers)
Q1 (25th percentile)
Median (50th percentile)
Q3 (75th percentile)
Maximum (not counting outliers)

```

```python
Q1 = df['height'].quantile(0.25)
Q3 = df['height'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['height'] < lower) | (df['height'] > upper)]
```

```python
print(upper)
print(lower)
```

```python
outliers
```

```python
#Using np.where() to mark outliers
df['is_outlier'] = np.where(
    (df['height'] < lower) | (df['height'] > upper),
    'Yes', 'No'
)
```

```python
df
```

```python
# Trimming means removing outliers completely from the dataset  not replacing them.
#trimming deletes the entire row containing the outlier.
new_df = df[(df['height'] > lower) & (df['height'] < upper)]
```

```python
new_df
```

```python
# Capping = replacing outliers with limits
df['height']=np.where(df['height']>upper,upper,np.where(df['height']<lower,lower,df['height']))
print(df)
```

```python
df['height'] = df['height'].clip(lower=lower, upper=upper)

#If a value is less than lower, it becomes lower
#If a value is greater than upper, it becomes upper
#Values within the range remain unchanged
```

```python
#Update category = 'tall' when:
c1 >= 3
gender == 'M'

df.loc[(df['c1'] >= 3) & (df['gender'] == 'M'), 'category'] = 'tall'
df.loc[(df['c1'] >= 3) | (df['salary'] > 50000), 'category'] = 'tall'
```
